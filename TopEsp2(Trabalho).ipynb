```python
""" Importação das bibliotecas """
import pandas as pd
import random as rd
import numpy as np
```

```python
""" preparação do arquivo csv """
#importando a planilha
data = pd.read_csv("br_inep_ideb_brasil.csv")
data
```

```python
# selecionando colunas com valores numericos
data = data.iloc[:, [0] + list(range(4, 11))]
data
```

```python
# Definindo o número de clusters
k = input("Numero de clusters:")
```

```python
#criando os centroids
centroids = []
for i in range(int(k)):
  centroids.append(data.iloc[rd.randint(0, len(data))].values.tolist())
```

```python
centroids
```

```python
for elemento in data.iloc[1:].values.tolist():
  for centroid in centroids:
    resultado = np.linalg.norm(np.array(elemento) - np.array(centroid))
    print(resultado)
# Realizar o cálculo da distância entre o elemento atual com cada
   #  um dos centróids, e associá-lo ao grupo do centróid com a menor distância.
# Definir os novos centróids através do cálculo da média entre todos os
# elementos do grupo.
```
